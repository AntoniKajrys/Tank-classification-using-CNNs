```python
import os
from PIL import Image
import pandas as pd
import random, shutil, re
from collections import Counter
source_dir = "."  
target_dir = "data_combined"  

splits = ["train", "test", "valid"]
exclude_folder = ["archive (1)","data_combined"]

for split in splits:
    split_dir = os.path.join(target_dir, split)
    os.makedirs(split_dir, exist_ok=True)
counter = 0
for class_name in os.listdir(source_dir):
    if class_name in exclude_folder or not os.path.isdir(class_name):
        continue

    class_path = os.path.join(source_dir, class_name)
    for split in splits:
        split_path = os.path.join(class_path, split)
        if not os.path.exists(split_path):
            continue
        label_file = None
        
        for f in os.listdir(split_path):
            if f.lower().endswith(".csv"):
                label_file = os.path.join(split_path, f)
                break

        if label_file:
            labe_pd = pd.read_csv(label_file)
        else:
            labe_pd = pd.DataFrame(columns=["filename", "class"])
        for root, dirs, files in os.walk(split_path):
            for file in files:
                if file.lower().endswith((".jpg", ".jpeg", ".png", ".bmp", ".gif")):
                    src = os.path.join(root, file)
                    matching = labe_pd.loc[labe_pd["filename"] == file, "class"]
                    if not matching.empty:
                        new_name = matching.values[0]
                        ext = os.path.splitext(file)[1]
                        print(new_name)
                        new_filename = f"{new_name}{ext}"
                    else:
                        new_filename = file  
                        print(f"File {file} not found in df")
                    counter += 1
                    dst = os.path.join(target_dir, split, f"{counter}_{new_filename}")
                    img = Image.open(src)
                    img = img.convert("RGB")             
                    img = img.resize((256, 256))         
                    img.save(dst)
                    
```

```python

base_dir = "data_combined"

# regex to catch t90, t-90, t_90, ...
t_re = re.compile(r"t[-_ ]?\d+", re.IGNORECASE)
# After running the next cell, russian tank were underrepresented in train set, so we move them
for src_split in ["test", "valid"]:
    src_path = os.path.join(base_dir, src_split)
    dst_path = os.path.join(base_dir, "train")


    files = []
    for f in os.listdir(src_path):
        name = os.path.splitext(f)[0].lower()
        if f.lower().endswith((".jpg", ".jpeg", ".png", ".bmp", ".gif")) and t_re.search(name):
            files.append(f)
        print(f"{src_split}: found {len(files)} TXX images")

    n = min(100, len(files))
    selected = random.sample(files, n)

    for f in selected:
        shutil.move(os.path.join(src_path, f), os.path.join(dst_path, f))

    print(f"moved {n} from {src_split} to= train")
```

```python

counts = Counter()

for split in splits:
    split_counts = Counter()
    split_path = os.path.join(base_dir, split)
    if not os.path.isdir(split_path):
        continue
    american_idx = 1
    israeli_idx = 1
    russian_idx = 1
    for fn in os.listdir(split_path):
        if not fn.lower().endswith((".jpg", ".jpeg", ".png", ".bmp", ".gif")):
            continue
        name = os.path.splitext(fn)[0].lower()
        file_path = os.path.join(split_path, fn)
        if "abrams" in name or "bradley" in name or "buford" in name:
            split_counts["Abrams"] += 1
            new_name = f"american_{american_idx}{ext}"
            american_idx += 1
            new_path = os.path.join(split_path, new_name)
            os.rename(file_path, new_path)
        elif "merkav" in name:
            split_counts["Merkavas"] += 1
            new_name = f"israeli_{israeli_idx}{ext}"
            israeli_idx += 1
            new_path = os.path.join(split_path, new_name)
            os.rename(file_path, new_path)
        elif t_re.search(name):
            split_counts["TXX"] += 1
            new_name = f"russian_{russian_idx}{ext}"
            russian_idx += 1
            new_path = os.path.join(split_path, new_name)
            os.rename(file_path, new_path)
        else:
            split_counts["Other"] += 1
            os.remove(file_path)
    counts.update(split_counts)

    print(f"\n{split.upper()}:")
    for k, v in split_counts.items():
        print(f"  {k}: {v}")

print("\nTOTAL:")
for k, v in counts.items():
    print(f"  {k}: {v}")
```
